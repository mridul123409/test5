### 1.Breakdown of Functions:
Mean: The mean is the sum of the list divided by the number of elements in the list.
Median: The median is the middle element when the list is sorted. If the list has an even length, the median is the average of the two middle elements.
Mode: The mode is the most frequent number in the list. If there are multiple modes, Python's statistics.mode will return the first one it encounters.
Weighted Mean: This is the mean where each number has an associated weight. The weighted mean is the sum of each value multiplied by its corresponding weight, divided by the total sum of weights.
Geometric Mean: The geometric mean is the nth root of the product of n numbers. It's useful when dealing with numbers that span several orders of magnitude or for multiplicative processes.

```python
import random
import statistics
import math

# (i) Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Calculate the mean of the list
def mean(lst):
    return sum(lst) / len(lst)

# (ii) Calculate the median of the list
def median(lst):
    return statistics.median(lst)

# (iii) Calculate the mode of the list
def mode(lst):
    return statistics.mode(lst)

# (iv) Calculate the weighted mean of the list
def weighted_mean(values, weights):
    return sum(v * w for v, w in zip(values, weights)) / sum(weights)

# (v) Calculate the geometric mean of the list (positive numbers only)
def geometric_mean(lst):
    product = 1
    for num in lst:
        if num <= 0:
            raise ValueError("Geometric mean is undefined for non-positive numbers.")
        product *= num
    return product ** (1 / len(lst))

# (vi) Calculate the harmonic mean of the list
def harmonic_mean(lst):
    return len(lst) / sum(1 / num for num in lst)

# (vii) Calculate the midrange (average of the minimum and maximum)
def midrange(lst):
    return (min(lst) + max(lst)) / 2

# (viii) Calculate the trimmed mean (excluding a certain percentage of outliers)
def trimmed_mean(lst, trim_percentage):
    sorted_lst = sorted(lst)
    n = len(lst)
    trim_count = int(n * trim_percentage / 100)
    trimmed_lst = sorted_lst[trim_count:-trim_count] if trim_count > 0 else sorted_lst
    return sum(trimmed_lst) / len(trimmed_lst)

# Examples of using the functions
print("Mean:", mean(int_list))
print("Median:", median(int_list))
print("Mode:", mode(int_list))
print("Geometric Mean:", geometric_mean([x for x in int_list if x > 0]))  # Assuming no non-positive numbers
print("Harmonic Mean:", harmonic_mean(int_list))
print("Midrange:", midrange(int_list))

# Trimmed mean with 10% trimming
trim_percentage = 10
print(f"Trimmed Mean (removing {trim_percentage}% outliers):", trimmed_mean(int_list, trim_percentage))

# Example of weighted mean (using random weights for demonstration)
weights = [random.randint(1, 5) for _ in range(100)]
print("Weighted Mean:", weighted_mean(int_list, weights))
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# (i) Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Visualization for the given data:
# 1. Frequency & Gaussian distribution
plt.figure(figsize=(12, 6))
sns.histplot(int_list2, kde=False, color='skyblue', bins=20, stat='density')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, np.mean(int_list2), np.std(int_list2))
plt.plot(x, p, 'k', linewidth=2)
plt.title("Frequency & Gaussian Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# 2. Frequency smoothened KDE plot
plt.figure(figsize=(12, 6))
sns.kdeplot(int_list2, shade=True, color='blue', bw_adjust=0.5)
plt.title("Frequency Smoothened KDE Plot")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# 3. Gaussian distribution & smoothened KDE plot
plt.figure(figsize=(12, 6))
sns.histplot(int_list2, kde=True, color='lightgreen', bins=20, stat='density')
plt.title("Gaussian Distribution & Smoothened KDE Plot")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# (ii) Calculate the range of the given list
def data_range(lst):
    return max(lst) - min(lst)

# (iii) Calculate the variance and standard deviation
def variance(lst):
    return np.var(lst)

def standard_deviation(lst):
    return np.std(lst)

# (iv) Calculate the interquartile range (IQR)
def interquartile_range(lst):
    Q1 = np.percentile(lst, 25)
    Q3 = np.percentile(lst, 75)
    return Q3 - Q1

# (v) Calculate the coefficient of variation
def coefficient_of_variation(lst):
    return np.std(lst) / np.mean(lst)

# (vi) Calculate the mean absolute deviation (MAD)
def mean_absolute_deviation(lst):
    mean_val = np.mean(lst)
    return np.mean([abs(x - mean_val) for x in lst])

# (vii) Calculate the quartile deviation
def quartile_deviation(lst):
    Q1 = np.percentile(lst, 25)
    Q3 = np.percentile(lst, 75)
    return (Q3 - Q1) / 2

# (viii) Calculate the range-based coefficient of dispersion
def range_based_dispersion(lst):
    return (np.max(lst) - np.min(lst)) / np.mean(lst)

# Example Calculations:
print("Range of the list:", data_range(int_list2))
print("Variance of the list:", variance(int_list2))
print("Standard Deviation of the list:", standard_deviation(int_list2))
print("Interquartile Range of the list:", interquartile_range(int_list2))
print("Coefficient of Variation of the list:", coefficient_of_variation(int_list2))
print("Mean Absolute Deviation (MAD) of the list:", mean_absolute_deviation(int_list2))
print("Quartile Deviation of the list:", quartile_deviation(int_list2))
print("Range-based Coefficient of Dispersion:", range_based_dispersion(int_list2))
```

### 3The class includes methods to calculate its expected value (mean) and variance.
Explanation:
Constructor (__init__):

Takes two parameters:
values: A list representing the possible outcomes of the discrete random variable.
probabilities: A list of probabilities corresponding to each value.
The constructor checks if:
The lengths of the values and probabilities lists match.
The probabilities are between 0 and 1 and sum to 1 (a basic requirement for a probability distribution).
Expected Value (expected_value):




```python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """
        Initializes the discrete random variable.
        
        :param values: List of possible values of the random variable
        :param probabilities: List of corresponding probabilities for each value
        """
        if len(values) != len(probabilities):
            raise ValueError("The number of values and probabilities must be the same.")
        if not all(0 <= p <= 1 for p in probabilities) or sum(probabilities) != 1:
            raise ValueError("Probabilities must be between 0 and 1, and must sum to 1.")
        
        self.values = values
        self.probabilities = probabilities
    
    def expected_value(self):
        """
        Calculates the expected value (mean) of the discrete random variable.
        
        E(X) = sum(x * P(x)), where x is a value and P(x) is its probability.
        
        :return: The expected value of the random variable.
        """
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))
    
    def variance(self):
        """
        Calculates the variance of the discrete random variable.
        
        Var(X) = E(X^2) - (E(X))^2
        
        :return: The variance of the random variable.
        """
        # Calculate E(X^2)
        expected_value_squared = sum(value**2 * prob for value, prob in zip(self.values, self.probabilities))
        
        # Calculate E(X)
        expected_value = self.expected_value()
        
        # Variance formula: Var(X) = E(X^2) - (E(X))^2
        return expected_value_squared - expected_value**2

# Example usage
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

# Create a DiscreteRandomVariable object
random_variable = DiscreteRandomVariable(values, probabilities)

# Calculate expected value and variance
print("Expected Value (E[X]):", random_variable.expected_value())
print("Variance (Var[X]):", random_variable.variance())
```

```python
import random
import numpy as np

# Function to simulate rolling a fair six-sided die
def roll_die():
    return random.randint(1, 6)

# Function to calculate expected value of the die roll
def expected_value():
    # For a fair six-sided die, the expected value is:
    return (1 + 2 + 3 + 4 + 5 + 6) / 6

# Function to calculate variance of the die roll
def variance():
    # The variance formula is Var(X) = E(X^2) - (E(X))^2
    E_X_squared = (1**2 + 2**2 + 3**2 + 4**2 + 5**2 + 6**2) / 6
    E_X = expected_value()
    return E_X_squared - E_X**2

# Simulate a large number of rolls to estimate expected value and variance
def simulate_rolls(num_rolls):
    rolls = [roll_die() for _ in range(num_rolls)]
    
    # Calculate empirical expected value (mean of the rolls)
    empirical_expected_value = np.mean(rolls)
    
    # Calculate empirical variance
    empirical_variance = np.var(rolls)
    
    return empirical_expected_value, empirical_variance

# Main program
if __name__ == "__main__":
    # Calculate the theoretical expected value and variance
    theoretical_E = expected_value()
    theoretical_Var = variance()
    
    print(f"Theoretical Expected Value: {theoretical_E}")
    print(f"Theoretical Variance: {theoretical_Var}")
    
    # Simulate 10000 die rolls to estimate the empirical expected value and variance
    num_rolls = 10000
    empirical_E, empirical_Var = simulate_rolls(num_rolls)
    
    print(f"Empirical Expected Value (from simulation of {num_rolls} rolls): {empirical_E}")
    print(f"Empirical Variance (from simulation of {num_rolls} rolls): {empirical_Var}")
```

```python
 import numpy as np

# Function to generate random samples and calculate mean and variance
def generate_samples(dist_type, *params, num_samples=1000):
    """
    Generate random samples from a given probability distribution and calculate their mean and variance.
    
    :param dist_type: The type of the distribution ('binomial' or 'poisson')
    :param params: Parameters for the distribution (e.g., n, p for binomial, lambda for Poisson)
    :param num_samples: The number of random samples to generate
    :return: A tuple (mean, variance) of the generated samples
    """
    # Generate random samples based on the distribution type
    if dist_type == 'binomial':
        if len(params) != 2:
            raise ValueError("Binomial distribution requires two parameters: n (trials) and p (probability).")
        n, p = params
        samples = np.random.binomial(n, p, num_samples)
    
    elif dist_type == 'poisson':
        if len(params) != 1:
            raise ValueError("Poisson distribution requires one parameter: lambda (rate).")
        lam = params[0]
        samples = np.random.poisson(lam, num_samples)
    
    else:
        raise ValueError("Unsupported distribution type. Choose either 'binomial' or 'poisson'.")
    
    # Calculate mean and variance of the generated samples
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    
    return sample_mean, sample_variance

# Example Usage
if __name__ == "__main__":
    # Binomial distribution example (n=10 trials, p=0.5 probability of success)
    binomial_mean, binomial_variance = generate_samples('binomial', 10, 0.5, num_samples=10000)
    print(f"Binomial Distribution - Mean: {binomial_mean}, Variance: {binomial_variance}")
    
    # Poisson distribution example (lambda=5, average rate of occurrence)
    poisson_mean, poisson_variance = generate_samples('poisson', 5, num_samples=10000)
    print(f"Poisson Distribution - Mean: {poisson_mean}, Variance: {poisson_variance}")
```

Steps:

Compute the mean of the generated samples using numpy.mean().

```python
import numpy as np

# Function to generate samples from a Gaussian distribution and calculate mean, variance, and standard deviation
def generate_gaussian_samples(mean, std_dev, num_samples=1000):
    """
    Generate random samples from a Gaussian distribution and compute their mean, variance, and standard deviation.
    
    :param mean: The mean (mu) of the Gaussian distribution.
    :param std_dev: The standard deviation (sigma) of the Gaussian distribution.
    :param num_samples: The number of samples to generate (default is 1000).
    :return: A tuple containing (mean, variance, standard deviation) of the samples.
    """
    # Generate random samples from a Gaussian distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
    
    # Calculate the mean, variance, and standard deviation
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev

# Example Usage
if __name__ == "__main__":
    # Parameters for the Gaussian distribution (mean = 0, std_dev = 1)
    mean = 0
    std_dev = 1
    num_samples = 10000
    
    # Generate samples and compute mean, variance, and standard deviation
    sample_mean, sample_variance, sample_std_dev = generate_gaussian_samples(mean, std_dev, num_samples)
    
    # Output the results
    print(f"Generated {num_samples} samples from a Gaussian distribution with mean = {mean} and std_dev = {std_dev}")
    print(f"Sample Mean: {sample_mean}")
    print(f"Sample Variance: {sample_variance}")
    print(f"Sample Standard Deviation: {sample_std_dev}")
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew, pearsonr

# Load the tips dataset
tips = sns.load_dataset('tips')

# (i) Function to calculate skewness
def calculate_skewness(data):
    """
    Calculate the skewness of a given numerical dataset.
    
    :param data: The dataset (a Pandas Series or list of numerical values).
    :return: The skewness value.
    """
    return skew(data)

# (ii) Program to determine skewness type
def skewness_type(data):
    """
    Determines whether a dataset exhibits positive, negative, or no skew.
    
    :param data: The dataset (a Pandas Series or list of numerical values).
    :return: A string indicating the type of skew ('Positive', 'Negative', or 'Approximately Symmetric').
    """
    skew_value = calculate_skewness(data)
    if skew_value > 0:
        return 'Positive Skew'
    elif skew_value < 0:
        return 'Negative Skew'
    else:
        return 'Approximately Symmetric'

# (iii) Function to calculate covariance between two columns
def calculate_covariance(col1, col2):
    """
    Calculate the covariance between two columns of a DataFrame.
    
    :param col1: First column (Pandas Series).
    :param col2: Second column (Pandas Series).
    :return: The covariance between the two columns.
    """
    return np.cov(col1, col2)[0][1]

# (iv) Function to calculate Pearson correlation coefficient
def calculate_pearson_correlation(col1, col2):
    """
    Calculate the Pearson correlation coefficient between two columns.
    
    :param col1: First column (Pandas Series).
    :param col2: Second column (Pandas Series).
    :return: Pearson correlation coefficient.
    """
    return pearsonr(col1, col2)[0]

# (v) Function to visualize the correlation between two columns with scatter plot
def plot_scatter_plot(col1, col2, df, title='Scatter Plot'):
    """
    Create a scatter plot to visualize the relationship between two columns.
    
    :param col1: First column name (string).
    :param col2: Second column name (string).
    :param df: DataFrame containing the columns.
    :param title: Title of the plot (default is 'Scatter Plot').
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(df[col1], df[col2], alpha=0.7)
    plt.title(title)
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()

# --- Now let's run these functions on 'total_bill' and 'tip' columns of the tips dataset ---

# (i) Calculate skewness of 'total_bill' and 'tip'
total_bill_skew = calculate_skewness(tips['total_bill'])
tip_skew = calculate_skewness(tips['tip'])

print(f"Skewness of total_bill: {total_bill_skew:.3f}")
print(f"Skewness of tip: {tip_skew:.3f}")

# (ii) Determine the skewness type
print(f"Skewness of total_bill: {skewness_type(tips['total_bill'])}")
print(f"Skewness of tip: {skewness_type(tips['tip'])}")

# (iii) Calculate covariance between 'total_bill' and 'tip'
covariance = calculate_covariance(tips['total_bill'], tips['tip'])
print(f"Covariance between 'total_bill' and 'tip': {covariance:.3f}")

# (iv) Calculate Pearson correlation coefficient between 'total_bill' and 'tip'
pearson_corr = calculate_pearson_correlation(tips['total_bill'], tips['tip'])
print(f"Pearson correlation coefficient between 'total_bill' and 'tip': {pearson_corr:.3f}")

# (v) Scatter plot to visualize correlation between 'total_bill' and 'tip'
plot_scatter_plot('total_bill', 'tip', tips, title='Scatter Plot: Total Bill vs Tip')
```

### 11To determine whether the new website layout leads to a higher conversion rate than the old layout, we can perform a two-proportion z-test. This statistical test compares two proportions (in this case, the conversion rates from the old and new layouts) to determine if there is a significant difference between them.

```python
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats

# Data
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Calculate the conversion rates for both layouts
p1 = np.mean(old_layout)  # Conversion rate for old layout
p2 = np.mean(new_layout)  # Conversion rate for new layout

# Number of visitors
n1 = len(old_layout)
n2 = len(new_layout)

# Number of successes (purchases)
X1 = np.sum(old_layout)
X2 = np.sum(new_layout)

# Pooled proportion
P = (X1 + X2) / (n1 + n2)

# Z-score calculation
Z = (p1 - p2) / np.sqrt(P * (1 - P) * (1/n1 + 1/n2))

# Calculate the p-value for a one-tailed test
p_value = stats.norm.cdf(Z)

# Display the results
print(f"Conversion rate for old layout (p1): {p1:.4f}")
print(f"Conversion rate for new layout (p2): {p2:.4f}")
print(f"Z-value: {Z:.4f}")
print(f"P-value: {p_value:.4f}")

# Set the significance level (alpha)
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject the null hypothesis: The new layout is significantly better.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

### 15 To perform an A/B test analysis comparing the click-through rates of two website layouts, we can use a two-sample t-test. The test will help us determine if there is a statistically significant difference between the two layouts in terms of the number of clicks.

```python
import numpy as np
import scipy.stats as stats

# Data: Number of clicks for each layout
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Step 1: Calculate sample means and standard deviations
mean_a = np.mean(layout_a_clicks)
mean_b = np.mean(layout_b_clicks)
std_a = np.std(layout_a_clicks, ddof=1)  # Sample standard deviation
std_b = np.std(layout_b_clicks, ddof=1)  # Sample standard deviation
n_a = len(layout_a_clicks)  # Sample size for Layout A
n_b = len(layout_b_clicks)  # Sample size for Layout B

# Step 2: Calculate the t-statistic
pooled_std = np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))  # Pooled standard error
t_statistic = (mean_a - mean_b) / pooled_std

# Step 3: Calculate degrees of freedom (using Welch-Satterthwaite equation)
df = ((std_a**2 / n_a + std_b**2 / n_b) ** 2) / (( (std_a**2 / n_a) ** 2) / (n_a - 1) + ((std_b**2 / n_b) ** 2) / (n_b - 1))

# Step 4: Compute the p-value for a two-tailed test
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df))

# Output the results
print(f"Mean of Layout A: {mean_a:.4f}")
print(f"Mean of Layout B: {mean_b:.4f}")
print(f"Standard deviation of Layout A: {std_a:.4f}")
print(f"Standard deviation of Layout B: {std_b:.4f}")
print(f"T-statistic: {t_statistic:.4f}")
print(f"Degrees of freedom: {df:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Decision based on significance level (alpha)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the click-through rates of the two layouts.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the click-through rates of the two layouts.")
```

### 19 To analyze whether there is a significant difference in quality between the two product versions based on the quality scores, we can perform a two-sample t-test. This statistical test compares the means of two independent samples (the quality scores for the two versions of the product) to determine if there is a statistically significant difference between them.

```python
import numpy as np
from scipy import stats

def t_test_version_quality(version1_scores, version2_scores, alpha=0.05):
    # Convert lists to numpy arrays for easier calculations
    version1_scores = np.array(version1_scores)
    version2_scores = np.array(version2_scores)
    
    # Calculate sample means and standard deviations
    mean1 = np.mean(version1_scores)
    mean2 = np.mean(version2_scores)
    std1 = np.std(version1_scores, ddof=1)  # Sample standard deviation
    std2 = np.std(version2_scores, ddof=1)
    n1 = len(version1_scores)
    n2 = len(version2_scores)
    
    # Calculate the t-statistic
    t_stat = (mean1 - mean2) / np.sqrt((std1**2 / n1) + (std2**2 / n2))
    
    # Calculate degrees of freedom using the Welch-Satterthwaite equation
    numerator = ((std1**2 / n1) + (std2**2 / n2))**2
    denominator = ((std1**2 / n1)**2 / (n1 - 1)) + ((std2**2 / n2)**2 / (n2 - 1))
    df = numerator / denominator
    
    # Calculate the p-value for the two-tailed test
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))  # two-tailed p-value
    
    # Compare the p-value to alpha and decide whether to reject the null hypothesis
    if p_value < alpha:
        result = "Reject the null hypothesis: significant difference in quality between the two versions."
    else:
        result = "Fail to reject the null hypothesis: no significant difference in quality between the two versions."
    
    return t_stat, df, p_value, result

# Given quality scores for version 1 and version 2
version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]
version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]

# Call the function
t_stat, df, p_value, result = t_test_version_quality(version1_scores, version2_scores)

# Output results
print(f"T-statistic: {t_stat:.4f}")
print(f"Degrees of Freedom: {df:.4f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

To determine if there is a significant association between age groups and voter preferences, you can perform a Chi-Square Test of Independence. This test will help you assess whether the distribution of voter preferences (Candidate A or Candidate B) depends on the different age groups.

The Chi-Square Test of Independence tests the null hypothesis that there is no association between the two categorical variables (in this case, age groups and voter preferences). If the p-value from the test is less than a chosen significance level (typically 0.05), we can reject the null hypothesis and conclude that there is a significant association between the variables.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate data using the code provided
np.random.seed(0)

# Define the age groups and preferences
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a DataFrame
data = pd.DataFrame({
    'Age Group': age_groups,
    'Voter Preference': voter_preferences
})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Output the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)
print("P-Value:", p_value)

# Decision based on p-value
if p_value < 0.05:
    print("\nReject the null hypothesis: There is a significant association between age group and voter preference.")
else:
    print("\nFail to reject the null hypothesis: No significant association between age group and voter preference.")
```
